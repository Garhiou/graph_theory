```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
adj_liste_cliques = {
    'A': ['B', 'C'],
    'B': ['A', 'C'],
    'C': ['A', 'B', 'D', 'E'], 
    'D': ['C', 'E'],
    'E': ['C', 'D'],
    'F': ['E'] 
}
```

```python
G = nx.from_dict_of_lists(adj_liste_cliques, create_using=nx.Graph)
```

```python
cliques = list(nx.find_cliques(G))
```

```python
three_cliques = [c for c in cliques if len(c) >= 3]
```

```python
print(f"Found cliques (size >= 3): {three_cliques}")
```

```python

plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G, seed=42) 


nx.draw_networkx_nodes(G, pos, node_color='lightgray', node_size=2000)


nx.draw_networkx_edges(G, pos, edge_color='gray')


nx.draw_networkx_labels(G, pos, font_size=15)


plt.title("Graph with two 3-cliques (A-B-C and C-D-E)")
plt.axis('off') 
plt.show()
```
